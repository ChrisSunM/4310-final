```python
biggerData = pd.read_csv("hd2021.csv", encoding= 'unicode_escape')
```

```python
smallerData = pd.read_csv("salary_potential.csv")
print(biggerData)
```

```python
finalcolleges = pd.DataFrame()
xplaceholder = 0
for x in biggerData.INSTNM:
  for y in smallerData.name:
    if x == y:
      finalcolleges = finalcolleges.append([biggerData.loc[xplaceholder]])
  xplaceholder += 1
print(len(finalcolleges))
print(finalcolleges)
finalcolleges = finalcolleges.reset_index()
print(finalcolleges)
```

```python
finalcolleges.to_csv("filtered_colleges.csv")
```

```python
incomeData = pd.read_csv("salary_potential.csv")
tuitionData = pd.read_csv("tuition_cost.csv")
```

```python
xplaceholder = 0
simplifiedData = pd.DataFrame()
for x in tuitionData.name:
  for y in incomeData.name:
    if x == y:
      simplifiedData = simplifiedData.append([tuitionData.loc[xplaceholder]])
  xplaceholder += 1

simplifiedData = simplifiedData.reset_index()
print(simplifiedData)
```

```python
xplaceholder = 0
finaldata = pd.DataFrame()
for x in simplifiedData.name:
  for y in finalcolleges.INSTNM:
    if x == y:
      finaldata = finaldata.append([simplifiedData.loc[xplaceholder]])
  
  xplaceholder += 1

finaldata = finaldata.reset_index()

finaldata = pd.merge(finaldata, incomeData[["name", "early_career_pay", "mid_career_pay", "make_world_better_percent", "stem_percent"]], on="name")
print(finaldata)
```

```python
finaldata.to_csv("TuitionIncome.csv")
```
